```python
syear = input("Enter year : ")
```

```python
stocks=input('Enter ticker: ')
```

```python
stocks=stocks.upper()
```

```python
# get the table for daily stock prices and,
# filter the table for selected tickers, columns within a time range
# set paginate to True because Quandl limits tables API to 10,000 rows per call
data = quandl.get_table('WIKI/PRICES', ticker = stocks, 
#                         date = { 'gte': '2015-12-31', 'lte': '2016-12-31' }, 
                        qopts = { 'columns': ['ticker', 'date', 'close'] }, 

                        paginate=True)

# create a new dataframe with 'date' column as index
new = data.set_index('date')
```

```python
# # use pandas pivot function to sort adj_close by tickers
df = new.pivot(columns='ticker')
```

```python
df.loc[syear].describe().round(decimals=2) #Bir baska yonden (istatistik) ozet
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

plt.figure(figsize=(15,6))

plt.plot(df.loc[syear])
# plt.grid()
plt.title('Stock Prices')
plt.ylabel('Dollar')
plt.xticks(rotation=45)
plt.legend(df)
plt.show()
```

```python
# We keep only one name in columns names. Remove 'close'.
df.columns = df.columns.get_level_values(1)
df.head()
```

```python
#Daily changes as (%)
df_pct = (df.pct_change()*100).round(decimals=2)
df_pct[syear].tail()
```

```python
#We merge daily prices dataframe with price changes (%) dataframe
#We add _% sign to tickers
df_merge = pd.merge(left=df, right=df_pct, on='date', suffixes=('', '_%')) #False, yeni isim verme demek.
df_merge[syear].tail()
```
